```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
xdata=iris.drop('species',axis=1)
ydata=iris.species
```

```python
lda=LDA(n_components=2)
```

```python
xdata=lda.fit_transform(xdata,ydata)
```

```python
sb.scatterplot(x=xdata[:,0],y=xdata[:,1],color='cyan')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,BaggingClassifier,AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
df=pd.DataFrame(xdata,columns=['pa1','pa2'])
```

```python
sb.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
```

```python
xtra,xtest,ytra,ytest=train_test_split(xdata,ydata,test_size=0.2)
```

```python
model1=LogisticRegression()
model1.fit(xtra,ytra)
model1_score=cross_val_score(model1,xdata,ydata,cv=7,verbose=1).mean()
```

```python
model2=DecisionTreeClassifier()
model2.fit(xtra,ytra)
model2_score=cross_val_score(model2,xdata,ydata,cv=7,verbose=1).mean()
```

```python
model3=BaggingClassifier()
model3.fit(xtra,ytra)
model3_score=cross_val_score(model3,xdata,ydata,cv=7,verbose=1).mean()
```

```python
model4=RandomForestClassifier()
model4.fit(xtra,ytra)
model4_score=cross_val_score(model4,xdata,ydata,cv=7,verbose=1).mean()
```

```python
model5=AdaBoostClassifier()
model5.fit(xtra,ytra)
model5_score=cross_val_score(model5,xdata,ydata,cv=7,verbose=1).mean()
```

```python
model6=XGBClassifier()
model6.fit(xtra,ytra)
model6_score=cross_val_score(model6,xdata,ydata,cv=7,verbose=1).mean()
```

```python
model7=MLPClassifier()
model7.fit(xtra,ytra)
model7_score=cross_val_score(model7,xdata,ydata,cv=7,verbose=1).mean()
```

```python
model8=SVC()
model8.fit(xtra,ytra)
model8_score=cross_val_score(model8,xdata,ydata,cv=7,verbose=1).mean()
```

```python
model9=GaussianNB()
model9.fit(xtra,ytra)
model9_score=cross_val_score(model9,xdata,ydata,cv=7,verbose=1).mean()
```

```python
model10=KNeighborsClassifier()
model10.fit(xtra,ytra)
model10_score=cross_val_score(model10,xdata,ydata,cv=7,verbose=1).mean()
```

```python
models=[ 'LogisticRegression',
 'DecisionTreeClassifier',
'RandomForestClassifier','BaggingClassifier','AdaBoostClassifier',
'XGBClassifier','MLPClassifier',
 'SVC',
'GaussianNB',
'KNeighborsClassifier']
scores=[
    model1_score,
    model2_score,
    model3_score,
    model4_score,
    model5_score,
    model6_score,
    model7_score,
    model8_score,
    model9_score,
    model10_score,
]
pd.DataFrame(data=dict(models=models,scores=scores))
```
